```python
import pandas as pd
import matplotlib.pyplot as plt
import glob

# Path to the directory with CSV files
dataset_dir = "secmatv1_csv"
csv_files = glob.glob(f"{dataset_dir}/*.csv")
len(csv_files)
```

```python
# Load the CSV file
file_path = csv_files[0]
data = pd.read_csv(file_path, header=None)  # No header in your CSV file

# Extract columns
x = data[0]  # Time values
y = data[1]  # Signal values

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(x, y, label="Signal")
plt.xlabel("Time (s)")
plt.ylabel("Signal Amplitude")
plt.title("DPA Contest Signal Data")
plt.legend()
plt.grid(True)
plt.show()
```
